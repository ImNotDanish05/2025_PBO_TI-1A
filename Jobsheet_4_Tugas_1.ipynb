```python
def bl():
  print("\n" * 1 + "=" * 50 + "\n" * 1)

class Mahasiswa:
  def __init__(self, nama, nim, ipk):
    self.nama = nama      # String
    self.nim = nim        # String
    self.ipk = float(ipk) # Float
  def __str__(self): #ini bisa berguna jika ngeprint variable saja
    return f"Nama: {self.nama}, NIM: {self.nim}, IPK: {self.ipk}"
  def __repr__(self): #ini bisa berguna jika diprint tanpa print() fungsi
    return f"Mahasiswa('{self.nama}', '{self.nim}', {self.ipk})"
  def tampilkan_info(self):
    bl()
    print(f"Nama: {self.nama}, NIM: {self.nim}, IPK: {self.ipk}")
  def hitung_predikat(self):
    ipk = self.ipk
    bl()
    if ipk >= 3.5:
      return "Dengan Pujian"
    elif ipk >= 3.0:
      return "Sangat Memuaskan"
    elif ipk >= 2.5:
      return "Memuaskan"
    elif ipk <= 2.5:
      return "Cukup"

class MahasiswaSarjana(Mahasiswa):
  def __init__(self, nama, nim, ipk, semester, sks_lulus):
    super().__init__(nama, nim, ipk)
    self.semester = semester
    self.sks_lulus = sks_lulus
  def tampilkan_info(self):
    super().tampilkan_info()
    print(f"Semester: {self.semester}, SKS Lulus: {self.sks_lulus}")
  def hitung_predikat(self):
    predikat = super().hitung_predikat()
    if self.semester <= 6 and self.sks_lulus >= 120:
      return f"{predikat} dan Lulus Cepat"
    else:
      return predikat

class MahasiswaMagister(Mahasiswa):
  def __init__(self, nama, nim, ipk, judul_tesis, nama_pembimbing):
    super().__init__(nama, nim, ipk)
    self.judul_tesis = judul_tesis
    self.nama_pembimbing = nama_pembimbing
  def tampilkan_info(self):
    super().tampilkan_info()
    print(f"Judul Tesis: {self.judul_tesis}, Nama Pembimbing: {self.nama_pembimbing}")
  def hitung_predikat(self):
    bl()
    if self.ipk >= 3.7:
      return "Dengan Pujian"
    elif self.ipk >= 3.3:
      return "Sangat Memuaskan"
    elif self.ipk >= 3.0:
      return "Memuaskan"
    else:
      return "Cukup"

Mahasiswa2 = MahasiswaSarjana("Jane Smith", "654321", 3.5, 8, 140)
Mahasiswa3 = MahasiswaMagister("Bob Johnson", "987654", 3.8, "Pengaruh Teknologi Informasi terhadap Pendidikan", "Dr. John Doe")
mahasiswa4 = MahasiswaSarjana("Alice Johnson", "123456", 2.5, 7, 130)
mahasiswa5 = MahasiswaMagister("Bob Smith", "654321", 2.3, "Pengaruh Teknologi Informasi terhadap Pendidikan", "Dr. Jane Doe")
mahasiswa6 = MahasiswaSarjana("Raka Smith", "123456", 3.5, 7, 130)

Mahasiswa2.tampilkan_info()
print("Predikat:", Mahasiswa2.hitung_predikat())

Mahasiswa3.tampilkan_info()
print("Predikat:", Mahasiswa3.hitung_predikat())

mahasiswa4.tampilkan_info()
print("Predikat:", mahasiswa4.hitung_predikat())

mahasiswa5.tampilkan_info()
print("Predikat:", mahasiswa5.hitung_predikat())

mahasiswa6.tampilkan_info()
print("Predikat:", mahasiswa6.hitung_predikat())
```
